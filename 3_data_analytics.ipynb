```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
# df_final = pd.read_csv("./data/dataset_balansed.csv")
df_final = pd.read_csv("./data/dataset_balansed_lema.csv")
```

```python
df_final.info()
```

```python
def graph_df(df):
    # Conteo de frecuencias
    counts = df["value"].value_counts().sort_index()

    # Gr치fico de barras
    plt.figure(figsize=(12,6))
    sns.barplot(x=counts.index, y=counts.values, palette="viridis")
    plt.title("Frecuencia de cada value")
    plt.xlabel("value")
    plt.ylabel("Cantidad de ejemplos")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Gr치fico de porcentaje
    percentages = counts / counts.sum() * 100
    plt.figure(figsize=(12,6))
    sns.barplot(x=percentages.index, y=percentages.values, palette="magma")
    plt.title("Porcentaje de cada value")
    plt.xlabel("value")
    plt.ylabel("Porcentaje (%)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Gr치fico de pastel (opcional, solo si hay pocas clases)
    if len(counts) <= 15:
        plt.figure(figsize=(8,8))
        plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
        plt.title("Distribuci칩n de value")
        plt.axis('equal')
        plt.show()
    
    # Tabla resumen
    summary = counts.reset_index()
    summary.columns = ['value', 'Cantidad']
    summary['Porcentaje'] = summary['Cantidad'] / summary['Cantidad'].sum() * 100

    print(summary)
```

```python
graph_df(df_final)
```
